```python
# importar o dataset dados_série_b
df = pd.read_csv('/content/gdrive/MyDrive/dados_serieB_2022_pontos.csv')
```

```python
# Definindo a coluna Rodadas como índice
df = df.set_index(['Rodadas'])
```

## Configurações da animação do gráfico

`df = df`: **Escolha dos dados que serão utilizados**

`title = Bar Chart Race - Brasileirão Série B - 2023`: **Título do gráfico**

`steps_per_period=10`: **Passos por período**

`period_length=500`: **Qtd de tempo por período definido para 500 milisegundos**

`figsize = (6, 3.5)`: **Tamanho da figura**

`orientation = ('h')`: **Orientação do gráfico, horizontal 'h' ou vertical 'v'**

```python
# Criação da animação do gráfico
bcr.bar_chart_race(df = df, title = 'Brasileirão Série B | 2022\nPontos dos Clubes',
                   steps_per_period = 10, period_length = 500,
                   figsize = (6, 3.5), orientation = 'h')
```

## Outras configurações de animação
Outras confgurações para deixar o seu gráfico mais estilizado conforme suas prêferências.


* Você pode salvar a animação em formato de vídeo .mp4 ou como arquvo html para exibição direto no navegar:


    filename='grafico_animado.mp4',  # insira aqui um nome + .mp4 ou .html para gerar o arquivo

* Organize a orientação do dados em modo ascendente ou descendente:


    sort='desc',  # orientar a organização dos dados: ascendente ou descendente

* Defina o tamanho máximo para 


    fixed_max=True,  # mantém o valor máximo fixo (por default é False)

* Dê um limite para a quantidade de barras do gráfico


    n_bars=20,  # limita a quantidade de barras no gráfico

* Mantenha a ordem fixa para as barras


    fixed_order=True,  # mantém uma ordem fixa das barras

* Defina do tamanho do texto do eixo x

    
    tick_label_size=7,  # tamanho do texto do eixo x
    
* Defina do tamanho do texto do eixo y


    bar_label_size=5,  # tamanho do texto do eixo y
    
* Ajuste o tamanho do título


    title_size='smaller',  # tamanho do título
    
* Interpolo de modo linear o valor do período

    
    interpolate_period=False,  # interpolar linearmente o valor do período
    
* Coloque texto ao lado das barras


    label_bars=True,  # colocar texto nas barras (por default é True)

* Controle o tamanho das barras

    
    bar_size=.95,  # tamanho das barras
* Ajuste a posição do texto dos períodos

    
    period_label={'x': .99, 'y': .25, 'ha': 'right', 'color': 'gray', 'va': 'center'},  # texto dos períodos
    
* Definição de cores


    cmap='dark24',  # mapa de cores
    
* Para reduzir a repetição de cores nas barras


    filter_column_colors=True)  # filtrar cores das colunas para reduzir repetição
